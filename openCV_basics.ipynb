## import required packages

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

## how to read image in opencv

```python
image = cv2.imread("photo-1501163268664-3fdf329d019f.jpeg")
```

```python
plt.imshow(image)
```

## Do you know why this image looks weird ? the main reason openCV read image in BGR color space by defalut so we need to convert it into RGB color space.

```python
# this is code for converting color space of an image.
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

# Now it looks wonderful 

```python
plt.imshow(RGB_image)
```

```python
# Here (500, 750, 3) it represents height, width and number of color channels (RGB) respectively.

RGB_image.shape
```

## Now convert this image into gray scale 

```python
gray_image = cv2.cvtColor(RGB_image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_image, cmap="gray")
```

## lets say we want to save this gray image in our local directory

```python
# Here we use cv2.imwrite to save this image

cv2.imwrite("gray_image.jpg", gray_image)

# if it return true thats mean your image written successfully.
```


## Now it is time to work on video data

```python
# Here we use pafy module to fetch video from youtube
import pafy

url = "https://www.youtube.com/watch?v=cKzP61Gjf00"
videoPafy = pafy.new(url)
best = videoPafy.getbest(preftype="webm")
print(videoPafy.duration)
print(videoPafy.description)
```

```python
video=cv2.VideoCapture('Most_Popular_Programming_Languages.mp4')

while(True):
    # Here ret is status and frame is image. (30 frame per second)
    ret, frame = video.read()
    if frame is not None:
        # Here I resized the pop-up size to 640*480 pixels
        resize = cv2.resize(frame, (640, 480))
        # cv2.imshow diplay frames here first argument is name of pop-up and resize is source
        cv2.imshow('Programing languages',resize)
        # Here we use wait key to hold video 
        # press q for quit this pop-up
        if cv2.waitKey(22) & 0xFF == ord('q'):
            break
    else:
        print("Frame is None")
        break


video.release()
cv2.destroyAllWindows()
print("Video stop")
```
